##Réduction de dimensionalité grace à l'ACP

```python
from sklearn import preprocessing
from sklearn import decomposition
```

```python
X = final_df[['sugars_100g', 'saturated-fat_100g', 'sodium_100g', 'fiber_100g', 'energy_100g', 'proteins_100g' ]]
```

```python
std_scale = preprocessing.StandardScaler().fit(X)
X_scaled = std_scale.transform(X)
```

```python
pca = decomposition.PCA(n_components=2)
pca.fit(X_scaled)
```

```python
X_projected = pca.transform(X_scaled)
sns.scatterplot(X_projected[:, 0], X_projected[:, 1], hue = final_df2['nutrition_grade_fr'] )
```

```python

pcs = pca.components_

for i, (x, y) in enumerate(zip(pcs[0, :], pcs[1, :])):
    plt.plot([0, x], [0, y], color='k')
    plt.text(x, y, X.columns[i], fontsize='14')

plt.plot([-1, 1], [0, 0], color='grey', ls='--')

plt.plot([0, 0], [-1, 1], color='grey', ls='--')

plt.xlim([-1, 1])
plt.ylim([-1, 1])
```

```python
display(pca.components_)
```

```python
print(pca.explained_variance_ratio_)
```
